# Ranking vs. Prize Money Per Point

For singles and doubles separately and each year individually, each player's ranking and their prize money earned divided number of points won is plotted to create a graph of ranking vs. prize money per point. These charts show the relationship between a player's ranking and how much money they earn per point they win.

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
```

```python
for year in range(START_YEAR, END_YEAR + 1):
    singles = pd.read_csv(f"../data/players/clean/singles/{year}.gz")

    # plot ranking x prize money/point
    plt.scatter(singles["ranking"], singles["prize_money"]/singles["points"])
    plt.xlabel("Ranking")
    plt.ylabel("Prize Money/Point")
    plt.title(f"Ranking x Prize Money/Point for Singles Players in {year}")
    plt.show()
```

```python
for year in range(START_YEAR, END_YEAR + 1):
    doubles = pd.read_csv(f"../data/players/clean/doubles/{year}.gz")

    # plot ranking x prize money/point
    plt.scatter(doubles["ranking"], doubles["prize_money"]/doubles["points"])
    plt.xlabel("Ranking")
    plt.ylabel("Prize Money/Point")
    plt.title(f"Ranking x Prize Money/Point for Doubles Players in {year}")
    plt.show()
```
