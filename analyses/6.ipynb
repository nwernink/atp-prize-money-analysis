# Correlation Between Ranking or Tour Level Matches Played and Prize Money Earned

The correlation between player rankings and prize money earned as well as tour level matches played and prize money earned is calculated by ranking band (1-50, 50-100, etc.) for each year for singles and doubles separately. The correlations illustrate how related the two variables are to each other.

```python
import pandas as pd
import datetime
import numpy as np
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
MAX_SINGLES = 750
MAX_DOUBLES = 250
```

```python
data = []
for year in range(START_YEAR, END_YEAR + 1):
    singles = pd.read_csv(f"../data/players/clean/singles/{year}.gz")

    band_prize_money = 0
    band_prize_money_array = []
    band_matches_played = 0
    band_matches_played_array = []
    band_ranking_array = []

    for rank in range(1, MAX_SINGLES + 1):
        player = singles.loc[singles["ranking"] == rank]
        if not player.empty:
            band_prize_money += player["prize_money"].iloc[0]
            band_prize_money_array.append(player["prize_money"].iloc[0])
            band_matches_played += player["wins"].iloc[0] + player["losses"].iloc[0]
            band_matches_played_array.append(player["wins"].iloc[0] + player["losses"].iloc[0])
            band_ranking_array.append(rank)

        if rank % 50 == 0:
            matches_correlation = np.corrcoef(band_prize_money_array, band_matches_played_array)[0, 1] ** 2
            ranking_correlation = np.corrcoef(band_prize_money_array, band_ranking_array)[1, 0] ** 2

            data.append({
                "year": year,
                "type": "singles",
                "ranking_band": f"{rank - 49} - {rank}",
                "prize_money": band_prize_money,
                "matches_played": band_matches_played,
                "matches_correlation": matches_correlation,
                "ranking_correlation": ranking_correlation
            })

            band_prize_money = 0
            band_prize_money_array = []
            band_matches_played = 0
            band_matches_played_array = []
            band_ranking_array = []


    doubles = pd.read_csv(f"../data/players/clean/doubles/{year}.gz")

    band_prize_money = 0
    band_prize_money_array = []
    band_matches_played = 0
    band_matches_played_array = []
    band_ranking_array = []

    for rank in range(1, MAX_DOUBLES + 1):
        player = doubles.loc[doubles["ranking"] == rank]
        if not player.empty:
            band_prize_money += player["prize_money"].iloc[0]
            band_prize_money_array.append(player["prize_money"].iloc[0])
            band_matches_played += player["wins"].iloc[0] + player["losses"].iloc[0]
            band_matches_played_array.append(player["wins"].iloc[0] + player["losses"].iloc[0])
            band_ranking_array.append(rank)

        if rank % 50 == 0:
            matches_correlation = np.corrcoef(band_prize_money_array, band_matches_played_array)[0, 1] ** 2
            ranking_correlation = np.corrcoef(band_prize_money_array, band_ranking_array)[1, 0] ** 2

            data.append({
                "year": year,
                "type": "doubles",
                "ranking_band": f"{rank - 49} - {rank}",
                "prize_money": band_prize_money,
                "matches_played": band_matches_played,
                "matches_correlation": matches_correlation,
                "ranking_correlation": ranking_correlation
            })

            band_prize_money = 0
            band_prize_money_array = []
            band_matches_played = 0
            band_matches_played_array = []
            band_ranking_array = []

res = pd.DataFrame(data)
```
