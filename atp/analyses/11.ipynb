# Average Prize Money Per Point at Each Round of Every Tournament

The average prize money per point at each round of each tournament level is calculated by averaging the prize money per point from each player at each round of each tournament by year.

```python
import pandas as pd
import datetime
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
TOURNAMENT_TYPES = ["GS", "1000", "500", "250", "CH"]
```

```python
df = pd.read_csv(f"../data/tournaments/tournaments_expanded.gz")
```

```python
singles = df[df["type"] == "singles"]
doubles = df[df["type"] == "doubles"]
```

```python
for tournament_type in TOURNAMENT_TYPES:
    tourney = singles[(singles["tournament_type"] == tournament_type)]

    rounds = tourney["round"].unique()
    for r in rounds:
        tourney_round = tourney[tourney["round"] == r] 
        print(tourney_round[["year", "tournament_type", "round", "average_prize_money_per_point"]])
```
