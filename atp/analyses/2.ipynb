# Ranking vs. Prize Money

For each year and for singles and doubles separately, each player's ranking and prize money earned is plotted to create a graph of ranking vs. prize money earned. These charts display how much prize money players earn in comparison to their rankings.

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
```

```python
for year in range(START_YEAR, END_YEAR + 1):
    singles = pd.read_csv(f"../data/players/clean/singles/{year}.gz")

    # plot ranking x prize money
    plt.scatter(singles["ranking"], singles["prize_money"])
    plt.xlabel("Ranking")
    plt.ylabel("Prize Money")
    plt.title(f"Ranking x Prize Money for Singles Players in {year}")
    plt.show()
```

```python
for year in range(START_YEAR, END_YEAR + 1):
    doubles = pd.read_csv(f"../data/players/clean/doubles/{year}.gz")

    # plot ranking x prize money
    plt.scatter(doubles["ranking"], doubles["prize_money"])
    plt.xlabel("Ranking")
    plt.ylabel("Prize Money")
    plt.title(f"Ranking x Prize Money for Doubles Players in {year}")
    plt.show()
```
