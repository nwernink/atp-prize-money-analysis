# Average Prize Money Per Point at Each Tournament Level

The average prize money per point at each tournament level is calculated by averaging the prize money per point from each player at each tournament level by year. The charts show the trend in how prize money per point has been awarded across the years to players as well as the differences in prize money per point awarded between different tournament levels.

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
START_YEAR = 2009
END_YEAR = datetime.datetime.now().year - 1
```

```python
df = pd.read_csv(f"../data/tournaments/tournaments_condensed.gz")
```

```python
singles = []
for year in range(START_YEAR, END_YEAR + 1):
    year_df = df[df["year"] == year]
    year_df = year_df[year_df["type"] == "singles"]

    gs, m, f, t, c = 0, 0, 0, 0, 0

    for _, row in year_df.iterrows():
        tournament_type = row["tournament_type"]

        if tournament_type == "Grand Slam":
            gs = row["average_prize_money_per_point"]
        elif tournament_type == "ATP Masters 1000":
            m = row["average_prize_money_per_point"]
        elif tournament_type == "ATP 500":
            f = row["average_prize_money_per_point"]
        elif tournament_type == "ATP 250":
            t = row["average_prize_money_per_point"]
        elif tournament_type == "Challenger":
            c = row["average_prize_money_per_point"]

    singles.append({
        "year": year,
        "grand_slam": gs,
        "masters_1000": m,
        "atp_500": f,
        "atp_250": t,
        "challenger": c,
    })

singles = pd.DataFrame(singles)
```

```python
plt.plot(singles["year"], singles["grand_slam"], label="Grand Slam")
plt.plot(singles["year"], singles["masters_1000"], label="Masters 1000")
plt.plot(singles["year"], singles["atp_500"], label="ATP 500")
plt.plot(singles["year"], singles["atp_250"], label="ATP 250")
plt.plot(singles["year"], singles["challenger"], label="Challenger")

plt.legend()
plt.title("Average Prize Money per Point by Tournament Level")
plt.xlabel("Year")
plt.ylabel("Average Prize Money per Point")
plt.show()
```
